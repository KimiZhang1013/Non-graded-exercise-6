Some typical NumPy 'random' functions:

```python
import numpy as np

# To generate a random float values between 0 and 1 (exclusive)
print(np.random.rand(3, 2)) # Creates a 3 x 2 matrix of random float values between 0 and 1 (exclusive)

# To generate random float values from standard normal distribution (mean = 0 and std = 1)
print(np.random.randn(5)) # Generates 5 random float values from standard normal distribution

# To generate random integer values within a range of values
print(np.random.randint(1, 100, 10)) # Generates 10 random integer values between 1 and 100 (exclusive)

# To randomly select an element from a given list of elements
print(np.random.choice([1, 2, 3, 4, 5, 6, 7, 8, 9], size=3)) # Three random numbers from the given array will be chosen

# To generate a random sample of values from a normally distributed with a specified mean and standard deviation
print(np.random.normal(loc=0, scale=1, size=10)) # 10 random values will be generated from a normally distributed data where 0 (loc) is the mean and 1 (scale) is the standard deviation

# To set a seed for a random number generation to ensure you get the same results later as well
print(np.random.seed(42)) # This ensures that any random numbers generated in the program are same every time you run the program. The seed value '42' can be any other value
```

```python
# Write your code for AIM #2 here

# Loading the CSV file as an array in NumPy
data = np.genfromtxt('sleep_health.csv', delimiter=',', skip_header=0, dtype=None, encoding='utf-8', names=True)

print(data.dtype.names)

# Create separate arrays for the individual columns that you want to operate on
start_time_pandas = time.time()
data_frame = pd.read_csv('sleep_health.csv')
mean_Sleep_Duration = data_frame['Sleep Duration'].mean()
print(mean_Sleep_Duration)
mean_Systolic_Blood_Pressure = data_frame['Systolic blood pressure'].mean()
print(mean_Systolic_Blood_Pressure)
mean_Diastolic_Blood_Pressure = data_frame['Diastolic blood pressure'].mean()
print(mean_Diastolic_Blood_Pressure)
mean_Heart_Rate = data_frame['Heart Rate'].mean()
print(mean_Heart_Rate)
mean_Daily_Steps = data_frame['Daily Steps'].mean()
print(mean_Daily_Steps)
elapsed_time_pandas = time.time() - start_time_pandas
print(f"Time taken by Pandas: {elapsed_time_pandas} seconds")

start_time_numpy = time.time()
Sleep_Duration = data['Sleep_Duration'].astype('float64')
mean_sleep_Duration = np.mean(Sleep_Duration)
print(mean_Sleep_Duration)
Systolic_Blood_Pressure = data['Systolic_blood_pressure'].astype('float64')
mean_Systolic_Blood_Pressure = np.mean(Systolic_Blood_Pressure)
print(mean_Systolic_Blood_Pressure)
Diastolic_Blood_Pressure = data['Diastolic_blood_pressure'].astype('float64')
mean_Diastolic_Blood_Pressure = np.mean(Diastolic_Blood_Pressure)
print(mean_Diastolic_Blood_Pressure)
Heart_Rate= data['Heart_Rate'].astype('float64')
mean_Heart_Rate = np.mean(Heart_Rate)
print(mean_Heart_Rate)
Daily_Steps = data['Daily_Steps'].astype('float64')
mean_Daily_Steps = np.mean(Daily_Steps)
print(mean_Daily_Steps)
elapsed_time_numpy = time.time() - start_time_numpy
print(f"Time taken by NumPy: {elapsed_time_numpy} seconds")

# Calculate the correlation in NumPy using np.corrcoef()
# The correlation coefficient is located in the off-diagonal elements [0,1] or [1,0]
start_time_pandas = time.time()
data_frame = pd.read_csv('sleep_health.csv')
correlation_age = data_frame['Age'].corr(data_frame['Sleep Duration'])
print(correlation_age)
correlation_heart = data_frame['Heart Rate'].corr(data_frame['Sleep Duration'])
print(correlation_heart)
correlation_steps = data_frame['Daily Steps'].corr(data_frame['Sleep Duration'])
print(correlation_steps)
print(f"Time taken by Pandas: {elapsed_time_pandas} seconds")

start_time_numpy = time.time()
correlation_age = np.corrcoef(data['Age'], data['Sleep_Duration'])[0,1]
print(correlation_age)
correlation_heart = np.corrcoef(data['Heart_Rate'], data['Sleep_Duration'])[0,1]
print(correlation_heart)
correlation_steps = np.corrcoef(data['Daily_Steps'], data['Sleep_Duration'])[0,1]
print(correlation_steps)
elapsed_time_numpy = time.time() - start_time_numpy
print(f"Time taken by NumPy: {elapsed_time_numpy} seconds")

# Use np-std() to calculate Standard deviation
start_time_pandas = time.time()
data_frame = pd.read_csv('sleep_health.csv')
std = data_frame['Sleep Duration'].std()
print(std)
print(f"Time taken by Pandas: {elapsed_time_pandas} seconds")

start_time_numpy = time.time()
std = np.std(data['Sleep_Duration'])
print(std)
elapsed_time_numpy = time.time() - start_time_numpy
print(f"Time taken by NumPy: {elapsed_time_numpy} seconds")
```

```python
# Write your code for AIM #3 here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data_frame = pd.read_csv('sleep_health.csv')

# Set the aesthetics for the plots
sns.set(style="whitegrid")

# 1.1: Age Distribution
data_frame['Age'].hist(bins=10, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# 1.2: Sleep Duration Distribution
data_frame['Sleep Duration'].hist(bins=10, color='lightgreen', edgecolor='black')
plt.title('Sleep Duration Distribution')
plt.xlabel('Sleep Duration (hours)')
plt.ylabel('Frequency')
plt.show()

# 1.3: Quality of Sleep Distribution
data_frame['Quality of Sleep'].hist(bins=10, color='salmon', edgecolor='black')
plt.title('Quality of Sleep Distribution')
plt.xlabel('Quality of Sleep (1-10)')
plt.ylabel('Frequency')
plt.show()

# 1.4: Physical Activity Level Distribution
data_frame['Physical Activity Level'].hist(bins=10, color='orange', edgecolor='black')
plt.title('Physical Activity Level Distribution')
plt.xlabel('Physical Activity Level')
plt.ylabel('Frequency')
plt.show()

# 1.5: Stress Level Distribution
data_frame['Stress Level'].hist(bins=10, color='purple', edgecolor='black')
plt.title('Stress Level Distribution')
plt.xlabel('Stress Level')
plt.ylabel('Frequency')
plt.show()

# 1.6: Heart Rate Distribution
data_frame['Heart Rate'].hist(bins=10, color='cyan', edgecolor='black')
plt.title('Heart Rate Distribution')
plt.xlabel('Heart Rate')
plt.ylabel('Frequency')
plt.show()

#To count unique values and their total count in a NumPy array, use 
# np.unique(array_name, return_counts=True)--- This will return two values, one is the unique values and the other is the total count of each unique value
# You will need this for plotting bar plots

# 3.1: Distribution of 'Sleep Duration' based on 'Quality of Sleep'
plt.figure(figsize=(12, 6))
sns.boxplot(x='Quality of Sleep', y='Sleep Duration', data=data_frame, palette='Set2')
plt.title('Sleep Duration Distribution by Quality of Sleep')
plt.xlabel('Quality of Sleep')
plt.ylabel('Sleep Duration (hours)')
plt.show()

# 3.2: Distribution of 'Sleep Duration' based on 'Stress Level'
plt.figure(figsize=(12, 6))
sns.boxplot(x='Stress Level', y='Sleep Duration', data=data_frame, palette='Set2')
plt.title('Sleep Duration Distribution by Stress Level')
plt.xlabel('Stress Level')
plt.ylabel('Sleep Duration (hours)')
plt.show()

# 3.3: Distribution of 'Sleep Duration' based on 'Physical Activity Level'
plt.figure(figsize=(12, 6))
sns.boxplot(x='Physical Activity Level', y='Sleep Duration', data=data_frame, palette='Set2')
plt.title('Sleep Duration Distribution by Physical Activity Level')
plt.xlabel('Physical Activity Level')
plt.ylabel('Sleep Duration (hours)')
plt.show()

# 3.4: Distribution of 'Sleep Duration' based on 'Occupation'
plt.figure(figsize=(12, 6))
sns.boxplot(x='Occupation', y='Sleep Duration', data=data_frame, palette='Set2')
plt.title('Sleep Duration Distribution by Occupation')
plt.xlabel('Occupation')
plt.ylabel('Sleep Duration (hours)')
plt.xticks(rotation=45)
plt.show()

# 3.5: Distribution of 'Sleep Duration' based on 'BMI Category'
plt.figure(figsize=(12, 6))
sns.boxplot(x='BMI Category', y='Sleep Duration', data=data_frame, palette='Set2')
plt.title('Sleep Duration Distribution by BMI Category')
plt.xlabel('BMI Category')
plt.ylabel('Sleep Duration (hours)')
plt.show()



# For box plots in NumPy you will need to 
# create a list of variable1, grouped by the unique value of some variable2 (nominal or ordinal), 
# Syntax: variable1_list = [variable1[variable2 == i] for i in np.unique(variable2)]
# 5.1: Scatter plot for Age and Sleep Duration
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Age', y='Sleep Duration', data=data_frame, color='blue')
plt.title('Scatter Plot of Age vs Sleep Duration')
plt.xlabel('Age')
plt.ylabel('Sleep Duration (hours)')
plt.show()

# 5.2: Scatter plot for Sleep Duration and Heart Rate
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Sleep Duration', y='Heart Rate', data=data_frame, color='green')
plt.title('Scatter Plot of Sleep Duration vs Heart Rate')
plt.xlabel('Sleep Duration (hours)')
plt.ylabel('Heart Rate (bpm)')
plt.show()

# 5.3: Scatter plot for Heart Rate and Daily Steps
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Heart Rate', y='Daily Steps', data=data_frame, color='orange')
plt.title('Scatter Plot of Heart Rate vs Daily Steps')
plt.xlabel('Heart Rate (bpm)')
plt.ylabel('Daily Steps')
plt.show()

# 5.4: Scatter plot for Sleep Duration and Daily Steps
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Sleep Duration', y='Daily Steps', data=data_frame, color='purple')
plt.title('Scatter Plot of Sleep Duration vs Daily Steps')
plt.xlabel('Sleep Duration (hours)')
plt.ylabel('Daily Steps')
plt.show()
```
